```python

def gaussian(t):
    total_sum = 0
    for i in range(0, 30):
        total_sum += np.exp((-((t - i*np.e)/np.e)**2)*np.log((t-np.e*i)/np.e))
    return total_sum
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def convolve_sequence(x,h):
    M = len(x)
    N = len(h)
    L = M+N-1
    y = [0]*L
    for n in range(L):
        for k in range(M):
            if n-k >=0 and n-k< N:
                y[n] = y[n] + (x[k]*h[n-k])
    return y
```

```python
def plot_sequence(x,h,y):
    nx = np.arange(0, len(x))
    nh = np.arange(0, len(h))
    ny = np.arange(0, len(y))
    plt.subplot(3,1,1)
    plt.stem(x)
    
    plt.subplot(3,1,2)
    plt.stem(h)
    
    plt.subplot(3,1,3)
    plt.stem(ny)
    
    plt.tight_layout()
    plt.show()
    
```

```python
x = [1,2,3]
h= [0.5,1]
y_c= convolve_sequence(x,h)
plot_sequence(x,h,y_c)
```

```python
def generate_signal(length, signal_type='sinusoidal'):
    if signal_type == 'sinusoidal':
        n = np.arange(length)
        x = np.sin(2 * np.pi * 0.1 * n)  
    else: 
        x = np.random.uniform(-2, 2, length) 
    return x
```

```python
def add_noise(signal, noise_level):
    noise = noise_level * np.random.normal(size=len(signal))
    noisy_signal = signal + noise
    return noisy_signal
```

```python
h1 = np.ones(3) / 3
h2 = np.ones(7) / 7
h3 = np.ones(11) / 11
```

```python
length_x = 10
x = generate_signal(length_x, signal_type='sinusoidal')
```

```python
noise_level = 0.1
x_noisy = add_noise(x, noise_level)
```

```python
y1 = convolve_sequence(x_noisy, h1)
y2 = convolve_sequence(x_noisy, h2)
y3 = convolve_sequence(x_noisy, h3)
```

```python
plt.plot(x_noisy)
plt.plot(y1)
plt.plot(y2)
plt.plot(y3)
plt.grid()
plt.show()
```
